```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
'''
Filter the training_file_all_seq in order to add a column that shows the positive/negative sequences.
'''
df =pd.read_csv('training_file_all_seq.tsv', sep='\t')
df['positive/negative'] =np.where(df['SP cleavage']=='False', '-','+')
```

```python
df
```

```python
df2 = pd.read_csv('benchmark_arricchito.tsv', sep='\t')
df2['subset_number'] = 'benchmark'
df2['positive/negative'] =np.where(df2['SP cleavage']=='False', '-','+')
```

```python
df2
```

**Comparison of Sequence Lengths Between Positive and Negative Classes**

```python
df_histograms_training = df[['Sequence length','positive/negative']]
df2_histograms_benchmark = df2[['Sequence length','positive/negative']]
```

```python
df_histograms_training
```

```python
df2_histograms_benchmark
```

```python
sns.histplot(
    data=df_histograms_training[df_histograms_training["Sequence length"]<=3000],
    x="Sequence length",
    hue="positive/negative",  
    common_norm = False, 
    stat="probability",      
)
plt.savefig("sequence_length_training_hist.png", dpi=300, bbox_inches="tight")
```

```python
sns.kdeplot(
    data=df_histograms_training[df_histograms_training["Sequence length"]<=3000],
    x="Sequence length",
    hue="positive/negative",  
    common_norm = False,       
)
plt.savefig("sequence_length_training_k.png", dpi=300, bbox_inches="tight")
```

```python
sns.histplot(
    data=df2_histograms_benchmark[df2_histograms_benchmark["Sequence length"]<=3000],
    x="Sequence length",
    hue="positive/negative",  
    common_norm = False, 
    stat="probability",      
)
plt.savefig("sequence_length_benchmark_hist.png", dpi=300, bbox_inches="tight")
```

```python
sns.kdeplot(
    data=df2_histograms_benchmark[df2_histograms_benchmark["Sequence length"]<=3000],
    x="Sequence length",
    hue="positive/negative",  
    common_norm = False,      
)
plt.savefig("sequence_length_benchmark_k.png", dpi=300, bbox_inches="tight")
```

```python
filter_positive_trainingset = df['positive/negative']== '+'
df_histograms_training_pos = df[filter_positive_trainingset]
dataframe_trainingset = df_histograms_training_pos[['Sequence length','subset_number']]
dataframe_trainingset['subset_number'] = 'trainingset'
```

```python
dataframe_trainingset
```

```python
filter_positive_benchmark = df2['positive/negative']== '+'
df_histograms_benchmark_pos = df2[filter_positive_benchmark]
dataframe_benchmark = df_histograms_benchmark_pos[['Sequence length','subset_number']]
```

```python
dataframe_benchmark
```

```python
#union of benchmark and training set
dataframe=pd.concat([dataframe_benchmark,dataframe_trainingset], ignore_index=True)
```

```python
sns.histplot(
    data=dataframe[dataframe["Sequence length"]<=3000],
    x="Sequence length",
    hue="subset_number",  
    common_norm = False, 
    stat="probability",      
)
plt.savefig("sequence_length_total_hist.png", dpi=300, bbox_inches="tight")
```

```python
sns.kdeplot( 
    data=dataframe[dataframe["Sequence length"]<=3000], 
    x="Sequence length", 
    hue="subset_number", 
    common_norm = False, 
    )
plt.savefig("sequence_length_total_k.png", dpi=300, bbox_inches="tight")
```
