```python
import pandas as pd
import numpy as np
```

```python
subset1 = pd.read_csv('subset1_arricchito.tsv', sep='\t')
subset2 = pd.read_csv('subset2_arricchito.tsv', sep='\t')
subset3 = pd.read_csv('subset3_arricchito.tsv', sep='\t')
subset4 = pd.read_csv('subset4_arricchito.tsv', sep='\t')
subset5 = pd.read_csv('subset5_arricchito.tsv', sep='\t')
lista = [subset1,subset2,subset3,subset4,subset5]
```

```python
subset1['positive/negative'] =np.where(subset1['SP cleavage']=='False', '-','+')
filter_positive_trainingset = subset1['positive/negative']== '+'
subset1_pos = subset1[filter_positive_trainingset]
subset2['positive/negative'] =np.where(subset2['SP cleavage']=='False', '-','+')
filter_positive_trainingset = subset2['positive/negative']== '+'
subset2_pos = subset2[filter_positive_trainingset]
subset3['positive/negative'] =np.where(subset3['SP cleavage']=='False', '-','+')
filter_positive_trainingset = subset3['positive/negative']== '+'
subset3_pos = subset3[filter_positive_trainingset]
subset4['positive/negative'] =np.where(subset4['SP cleavage']=='False', '-','+')
filter_positive_trainingset = subset4['positive/negative']== '+'
subset4_pos = subset4[filter_positive_trainingset]
subset5['positive/negative'] =np.where(subset5['SP cleavage']=='False', '-','+')
filter_positive_trainingset = subset5['positive/negative']== '+'
subset5_pos = subset5[filter_positive_trainingset]
```

```python
sub1 = subset1_pos['Sequence'].tolist()
sub2 = subset2_pos['Sequence'].tolist()
sub3 = subset3_pos['Sequence'].tolist()
sub4 = subset4_pos['Sequence'].tolist()
sub5 = subset5_pos['Sequence'].tolist()

def sp_extraction(subset):
    sp_90res=[]
    for i in subset:
        sp_90res.append(i[:90])
    return sp_90res
sp_extraction(sub1)
sp_extraction(sub2)
sp_extraction(sub3)
sp_extraction(sub4)
sp_extraction(sub5)

```

```python
df_sub1 = pd.DataFrame(sp_extraction(sub1), columns=['sp_seq_90'])
df_sub2 = pd.DataFrame(sp_extraction(sub2), columns=['sp_seq_90'])
df_sub3 = pd.DataFrame(sp_extraction(sub3), columns=['sp_seq_90'])
df_sub4 = pd.DataFrame(sp_extraction(sub4), columns=['sp_seq_90'])
df_sub5 = pd.DataFrame(sp_extraction(sub5), columns=['sp_seq_90'])
df_sub1['subset'] = 1
df_sub2['subset'] = 2
df_sub3['subset'] = 3
df_sub4['subset'] = 4
df_sub5['subset'] = 5
```

Position Specific Probability Matrix
INITALIZATION (PSPM): frequency of each residue type at each position

```python
aminoacids = ["A","R","N","D","C","Q","E","G","H","I",
              "L","K","M","F","P","S","T","W","Y","V"]
def init_matrix():
    positions = list(range(1, 91))
    matrix = pd.DataFrame(1, index=positions, columns=aminoacids)
    return matrix

```

```python
print(init_matrix())
```

FUNZIONE CHE SOMMA LE RICORRENZE PER OGNI POSIZIONE CON LA FUNZIONE ENUMERATE CHE PRENDE COME VARIABILE QUALUNQUE ITERABILE (LISTA, STRINGA...) CON DUE VARIABILI IN QUESTO CASO ABBIAMO UTILIZZATO LA POSIZIONE E LA COLONNA. SE L'AMINOACIDO VIENE TROVATO NELLA SEQUENZA VIENE INCREMENTATA DI UNO LA RISPETTIVA CELLA 

```python
def filling_matrix (sub,matrix):
    for seq in sub['sp_seq_90']:
        for pos, aa in enumerate(seq , start=1):
                if aa in matrix.columns:      
                    matrix.loc[pos, aa] += 1
    return matrix
```

```python
mat_sub1 = filling_matrix(df_sub1,init_matrix())
mat_sub2 = filling_matrix(df_sub2,init_matrix())
mat_sub3 = filling_matrix(df_sub3,init_matrix())
mat_sub4 = filling_matrix(df_sub4,init_matrix())
mat_sub5 = filling_matrix(df_sub5,init_matrix())

size_sub1=len(sub1)
size_sub2=len(sub2)
size_sub3=len(sub3)
size_sub4=len(sub4)
size_sub5=len(sub5)
```
